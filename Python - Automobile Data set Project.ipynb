```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV

df = pd.read_csv(r'C:\Users\GonDesktop\Documents\Github\PortfolioProjects\AutomobileDataProject.csv')
```

```python
#This allows you to see all the columns, not just like 15

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Columns datatype

df.info()
```

```python
# Drop columns we will not use

#df.drop(columns = 'highway-mpg', inplace = True)
#df = df.drop(columns = 'city-mpg')
df = df.drop(columns = 'stroke')
df = df.drop(columns = 'bore')
df = df.drop(columns = 'curb-weight')
df = df.drop(columns = 'drive-wheels')
df = df.drop(columns = 'aspiration')
df = df.drop(columns = 'symboling')
df = df.drop(columns = 'normalized-losses')
```

```python
# Dataset statistics

df.describe()
```

```python
# See how many NULLs we have in the dataset

df.isnull().sum()
```

```python
# See unique 

df.nunique()
```

```python
# Name change to make column

df = df.rename(columns = {'make' : 'brand'})
```

```python
# Columns with number data type

df.select_dtypes(include = 'number')
```

```python
# All columns

df.columns
```

```python
# Now we want to conver horsepower, peak-rpm and price columns into numeric values
# Convert to numeric, replacing errors with NaN

df['horsepower'] = pd.to_numeric(df['horsepower'], errors = 'coerce').astype('Int64')

# Convert multiple columns using pd.to_numeric with errors='coerce'

df[['peak-rpm','price','city-mpg']] = df[['peak-rpm','price','city-mpg']].apply(pd.to_numeric, errors = 'coerce').astype('Int64')
```

```python
# Select numeric type columns

numeric_df = df.select_dtypes('number')
```

```python
# Generate heatmap of numeric correlations

sns.heatmap(numeric_df.corr(), annot = True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
# Average price per brand

price_by_brand = df.groupby('brand')['price'].mean().sort_values(ascending = False)
plt.figure(figsize = (12,6))
sns.barplot(x = price_by_brand.index, y = price_by_brand.values, palette = 'magma')
plt.xticks(rotation = 90)
plt.xlabel('Brand')
plt.ylabel('Average Price (USD)')
plt.title('Car Average Price per Brand')
plt.show()
```

```python
# Iterate over the DataFrame row by row

for index, row in df.iterrows():
    # Check if the value in the 'brand' column is 'dodge'
    if row['brand'] == 'dodge':
        # Print the message with the brand and price of the car
        print(f"Car brand {row['brand']} is valued {row['price']}")
```

```python
# Potency and oil consumption relation

plt.figure(figsize = (12,6))
sns.scatterplot(x = df['horsepower'], y = df['city-mpg'], hue = df['brand'], alpha = 0.7, palette = 'Paired')
plt.legend(title = 'BranD', loc = 'upper right', fontsize = 7, title_fontsize = 9)
plt.xlabel('Potency (HP)')
plt.ylabel('City Consumption (MPG)')
plt.title('Potency and Oil Consumption Relation')
plt.show()
```

```python
# Category Car Segmentation

bins = [0, 10000, 20000, 40000, df['price'].max()]
labels = ['Cheap','Intermediate','Premium','Luxury']
df['segment'] = pd.cut(df['price'], bins = bins, labels = labels)

plt.figure(figsize = (8,5))
sns.countplot(x = df['segment'], order = labels, hue = df['segment'], palette = 'Dark2')
plt.xlabel('Car Segment')
plt.ylabel('Cars Quantity')
plt.title('Car Distribution by Price Segment')
plt.show()
```

```python
# Show cars by segment table

auto_segments = df.groupby('segment')['brand'].count()
print(auto_segments)
```

```python
# Compare city consumption vs highway consumption

plt.figure(figsize = (10,6))
sns.scatterplot(x = df['city-mpg'], y = df['highway-mpg'], hue = df['brand'], alpha = 0.7)
plt.legend(title = 'BranD', loc = 'upper right', fontsize = 7, title_fontsize = 9)
plt.xlabel('City Consumption (MPG)')
plt.ylabel('Highway consumption (MPG)')
plt.title('Car Consumption, City vs Highway')
plt.show()
```

```python
# Engine displacement distribution

if "engine-size" in df.columns:
    #df["engine-size"] = pd.to_numeric(df["engine-size"], errors="coerce")
    plt.figure(figsize = (8, 5))
    sns.histplot(df["engine-size"].dropna(), bins=20, kde=True)
    plt.xlabel("Motor Size (CC)")
    plt.ylabel("Frequency")
    plt.title("Motor Size Distribution")
    plt.show()
```

```python
# Dataset to csv

output_file = 'Automobile_CleanedProject.csv'
df.to_csv(output_file, index = False)
print(f'Cleaned Dataset saved as {output_file}')
```
